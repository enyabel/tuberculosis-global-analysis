La tuberculosis (TB) es una enfermedad infecciosa causada principalmente por la bacteria Mycobacterium tuberculosis, la cual afecta sobre todo a los pulmones, aunque también puede comprometer otros órganos como huesos, riñones o sistema nervioso. Se transmite por vía aérea cuando una persona infectada tose, estornuda o habla, liberando pequeñas gotas que contienen la bacteria. La tuberculosis ha acompañado a la humanidad desde hace miles de años; existen evidencias de su presencia en restos humanos de más de 9,000 años de antigüedad. Sin embargo, su agente causal fue identificado científicamente hasta 1882 por el médico alemán Robert Koch, quien descubrió el bacilo responsable de la enfermedad.

El diagnóstico de la tuberculosis combina evaluación clínica, estudios de imagen y pruebas de laboratorio. Entre los métodos más utilizados se encuentran la radiografía de tórax para detectar lesiones pulmonares sugestivas, la baciloscopía de esputo para identificar el bacilo en muestras respiratorias, y el cultivo microbiológico, considerado el estándar de confirmación diagnóstica. También se emplean pruebas moleculares rápidas como la PCR/Xpert MTB-RIF, que permiten detectar material genético de la bacteria e incluso resistencia a ciertos antibióticos en pocas horas. Para identificar infección latente se utilizan la prueba cutánea de tuberculina (PPD/Mantoux) y los análisis de sangre tipo IGRA.

**Limpieza y manipulación de datos**

```python
# Cuantos valores faltantes hay por columna?
who.isna().sum()
```

```python
# Explorar los valores faltantes en la columna iso2
who[who.iso2.isna()]
```

```python
# Los valores NA se debe a que el iso2 de Namibia es "NA", y pandas lo tomo como un valor faltante
who.loc[who.country == "Namibia", "iso2"] = "NA"
```

```python
# Las demas columnas de casos con valores faltantes van a tener un valor default de 0
who.fillna(0, inplace=True)
```

```python
# Cuantos valores falantes hoy por columna en la tabla de pop?
pop.isna().sum()
```

```python
who2 = who.melt(id_vars=["country", "iso2", "iso3", "year"])
who2.head(10)
```

```python
# Crear una columna que contenga el genero (Masculino o Femenino)
def asignar_genero(valor):
  if "m" in valor:
    return "Masculino"
  else:
    return "Femenino"
```

```python
who2["genero"] = who2["variable"].apply(asignar_genero)
who2.head(15)
```

```python
def asignar_metodo(valor):
  if "rel" in valor:
    return "relapse"
  elif "sn" in valor:
    return "negative pulmonary smear"
  elif "sp" in valor:
    return "positive pulmonary smear"
  elif "ep" in valor:
    return "extrapulmonary"
  else:
    return "to define"
```

```python
who2["metodo"] = who2["variable"].apply(asignar_metodo)
who2.head()
```

```python
def asignar_grupoedad(valor):
  if "014" in valor:
    return "0-14"
  elif "1524" in valor:
    return "15-24"
  elif "2534" in valor:
    return "25-34"
  elif "3544" in valor:
    return "35-44"
  elif "4554" in valor:
    return "45-54"
  elif "5564" in valor:
    return "55-64"
  else:
    return "65+"
```

```python
who2["grupoedad"] = who2["variable"].apply(asignar_grupoedad)
who2.head()
```

```python
who2.drop(columns=["variable"], inplace=True)
```

```python
who2.rename(columns={"value":"cases"}, inplace=True)
```

```python
who2.head()
```

```python
df = pd.merge(who2, pop, on=["country", "year"], how="inner")
df.head(10)
```

```python
df.to_csv("tuberculosis_class.csv")
df.to_csv("/content/drive/MyDrive/E/tuberculosis_class.csv")
```

```python
por_genero = df.groupby("genero", as_index = False)["cases"].sum()
por_genero["porcentaje"] = 100 * por_genero["cases"] / por_genero["cases"].sum()
por_genero.head()
```

```python
por_metodo = df.groupby("metodo", as_index=False)["cases"].sum()
por_metodo["porcentaje"] = 100 * por_metodo["cases"] / por_metodo["cases"].sum()
por_metodo.head()
```

**Tasa de incidencia (por cada 100 mil habitantes) = 100,000 * casos nuevo / población**

```python
country_year = df.groupby(["country", "year"], as_index=False).agg({"cases":"sum", "population":"max"})
country_year["tasa_incidencia"] = 100_000 * country_year["cases"] / country_year["population"]
country_year.head(10)
```

```python
by_year = country_year.groupby("year", as_index=False).agg(
    {"cases":"sum", "population": "sum"}
)
by_year["tasa_incidencia"] = 100_000 * by_year["cases"] / by_year["population"]
by_year
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="year", y="tasa_incidencia", data=by_year, color='red')
plt.title("Tasa de incidencia por año")
plt.xlabel("Año")
plt.ylabel("Tasa de incidencia")
# guardar la grafica en nuestro drive
plt.savefig("/content/drive/MyDrive/E/grafica_incidencia.png")
```
